```r
# 加载数据
tsData<-read.csv("D:\\projects\\datasets\\basexl.csv",header=T)
# 转换为时序数据
tsData<-ts(tsData$农业,start=tsData$年份[1])
n<-length(tsData)
plot.ts(tsData)
# 此处预留十年数据作为检验
orgPred=tsData[(n-9):n]
tsData=tsData[1:(n-10)]
```

```r
# 测试加载华为数据（半年）
tsData<-read.csv("..\\datasets\\weeksum.csv",header=T)
tsData$date=as.Date(tsData$date)
tsData<-ts(tsData$sale,start=tsData$date[1])
n<-length(tsData)
plot.ts(tsData)
# 此处预留十年数据作为检验
orgPred=tsData[(n-9):n]
tsData=tsData[1:(n-10)]
```

```r
# 进行d阶差分运算
d=1
z=NULL
for(t in (d+1):length(tsData))
{
    tmp=0
    for(i in 0:d)
    {
        tmp=tmp+(-1)^i*(factorial(d)/(factorial(i)*factorial(d-i)))*(tsData[t-i])
    }
    z=c(z,tmp)
}
# 使用单位根检验差分序列的平稳性
library(fUnitRoots)
unitrootTest(z)
```

```r
plot.ts(tsData)
```

```r
# 使用ljung-Box检验差分后的序列是否为白噪声
Box.test(z,type="Ljung-Box")
```

```r
# 分别绘制查分后虚了的自相关图和偏自相关图
par(mfrow=c(1,2))
acf(z)
pacf(z)
par(mfrow=c(1,1))
```
